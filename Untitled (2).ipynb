```python
#Implement Kepler's eqn using newtons method to find roots

def kepler(M,e):
    E0 = M
    E = E0
    
    for i in range(200):
        g = E0 - e*np.sin(E0) - M
        gp = 1 - e*np.cos(E0)
        E = E0 - g/gp
        
        if np.abs((E-E0)/E <=1e-11):
            return E
        E0=E
    return E
```

```python
a_earth = 149.6e6
e_earth = 0.0167
w_earth = 288.1*np.pi/180
P_earth = 365.256
```

```python
f = np.linspace(0, 2*np.pi,100)
r_earth = a_earth*(1-e_earth**2)/(1+e_earth*np.cos(f))

plt.plot(f, r_earth)
plt.xlabel('true anomoly')
plt.ylabel('r_earth km');
```

```python
plt.figure(figsize=(4,4))
plt.plot(r_earth*np.cos(f), r_earth*np.sin(f));
```

```python
n = 2*np.pi/P_earth
phi0 = 0
t0 = (phi0 + w_earth)/n

t = np.linspace(t0, t0+ 365, 100)
M = n*t - (phi0+ w_earth)
f = np.array([kepler(m, e_earth) for m in M])
r_earth = a_earth * (1-e_earth**2)/(1+e_earth*np.cos(f))

plt.plot(t,r_earth)
plt.xlabel('time')
plt.ylabel('r_earth');
```

```python
from astropy.constants import G, M_sun, M_earth
from astropy import units as u
G = G.to(u.km**3/(u.kg * u.s**2))
G
```

```python
i = np.pi/2
K = np.sqrt(G/(1 - e_earth**2)) * M_sun * np.sin(i) * 1/np.sqrt(M_earth + M_sun) * 1/np.sqrt(a_earth)
rv_earth = K * (np.cos(w_earth + f) + e_earth * np.cos(w_earth))

plt.plot(t, rv_earth)
plt.xlabel('time[d]')
plt.ylabel('radial velocity [m/s]');
```

```python
data = pd.read_csv('UID_0113357_RVC_001.tbl', delim_whitespace=True, header=None, skiprows=21, names=['t', 'rv', 'rv_err'])
data
```

```python
epoch = np.mean(data.t)
data.t -= epoch
data
```

```python
plt.errorbar(data.t, data.rv, yerr=data.rv_err, fmt='.k', capsize=0)
plt.xlabel('time [d]')
plt.ylabel('radial velocity [m/s]')
plt.xlim(-375,-365);
```

```python
from astropy.timeseries import LombScargle
```

```python
frequencies, power = LombScargle(data.t, data.rv).autopower()
```

```python
periods = 1/frequencies

plt.semilogx(periods, power)
plt.xlabel('period')
plt.ylabel('power');
```

```python
period = periods[np.argmax(power)]
period
```

```python
phase = (data.t % period)/period
```

```python
plt.errorbar(phase, data.rv, yerr=data.rv_err, fmt='.k', capsize=0)

plt.xlabel('phase')
plt.ylabel('radial velocity');
```

```python
import pymc3 as pm

from exoplanet.orbits import get_true_anomaly
from exoplanet.distributions import Angle
```

```python
with pm.Model() as model:
    logK = pm.Uniform('logK', lower=0, upper=np.log(200),
                      testval=np.log(0.5 * (np.max(data.rv) - np.min(data.rv))))
    logP = pm.Uniform('logP', lower=0, upper=np.log(10), testval=np.log(period))
    phi = pm.Uniform('phi', lower=0, upper=2*np.pi, testval=0.1)
    e = pm.Uniform('e', lower=0, upper=1, testval=0.1)
    w = Angle("w")
    logjitter = pm.Uniform('logjitter', lower=-10, upper=5, testval=np.log(np.mean(data.rv_err)))
    rv0 = pm.Normal('rv0', mu=0, sd=10, testval=0)
    rvtrend = pm.Normal('rvtrend', mu=0, sd=10, testval=0)
    
    n = 2 * np.pi * pm.math.exp(-logP)
    P = pm.Deterministic('P', pm.math.exp(logP))
    K = pm.Deterministic('K', pm.math.exp(logK))
    cosw = pm.math.cos(w)
    sinw = pm.math.sin(w)
    s2 = pm.math.exp(2 * logjitter)
    t0 = (phi + w) / n
    
    bkg = pm.Deterministic('bkg', rv0 + rvtrend * data.t / 365.25)
    M = n * data.t - (phi + w)
    
    f = get_true_anomaly(M, e + pm.math.zeros_like(M))
    rvmodel = pm.Deterministic("rvmodel",
                               bkg + K * (cosw * (pm.math.cos(f) + e) - sinw * pm.math.sin(f)))
    
    err = pm.math.sqrt(data.rv_err.to_numpy()**2 + pm.math.exp(2 * logjitter))
    pm.Normal('obs', mu=rvmodel, sd=err, observed=data.rv)
    
    phase = np.linspace(0, 1, 500)
    M_pred = 2 * np.pi * phase - (phi + w)
    f_pred = get_true_anomaly(M_pred, e + pm.math.zeros_like(M_pred))
    rvphase = pm.Deterministic('rvphase',
                               K * (cosw * (pm.math.cos(f_pred) + e) - sinw * pm.math.sin(f_pred)))
```

```python
with model:
    map_params = pm.find_MAP()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 8))

period = map_params['P']

ax = axes[0]
ax.errorbar(data.t, data.rv, yerr=data.rv_err, fmt='.k')
ax.plot(data.t, map_params['bkg'], color='C0', lw=1)
ax.set_ylim(-110, 110)
ax.set_ylabel('radial velocity [m/s]')
ax.set_xlabel('time [d]')

ax = axes[1]
ax.errorbar(data.t % period, data.rv - map_params['bkg'], yerr=data.rv_err, fmt='k.')
ax.plot(phase * period, map_params['rvphase'], color='C1', lw=1)
ax.set_ylim(-110, 110)
ax.set_ylabel('radial velocity [m/s]')
ax.set_xlabel('phase [days]')

plt.tight_layout()
```

```python
with model:
    trace = pm.sample(draws=1000, tune=1000, start=map_params)
```
