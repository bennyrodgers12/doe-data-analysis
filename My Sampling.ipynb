```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.distributions import norm
```

```python
mu = 10.
sigma = 2.
def p(x):
    return .5 * (norm.pdf(x, loc=mu, scale=sigma) + \
                norm.pdf(x, loc=mu+5, scale=sigma/2))
```

```python
x = np.linspace(-40, 40, 500)
plt.plot(x, p(x));
```

```python
def q(x):
    return norm.pdf(x, loc=10, scale=6)
```

```python
k = max(p(x)/q(x))
```

```python
plt.plot(x, p(x), label='$p(x)$')
plt.plot(x, k*q(x), label='$k * q(x)$')
plt.legend();
```

```python
def rejection_sample(iter=100):
    samples = []
    
    for i in range(iter):
        z = norm.rvs(loc=10, scale=6)
        u = np.random.uniform(0, k*q(z))
        
        if u <= p(z):
            samples.append(z)
    return np.array(samples)
```

```python
samps = rejection_sample(1000)
```
