```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats.distributions import beta
```

```python
theta_grid = np.linspace(0, 1, 1000)
```

```python
def plot_beta_dist(a=1, b=1, **kwarg):
    plt.plot(theta_grid, beta.pdf(theta_grid, a, b))
    plt.xlim(0,1)
    plt.ylim(0,3)
    plt.xlabel(r'$\theta$')
    plt.ylabel(r'$p(\theta|a={}, b={})$'.format(a,b))
```

```python
plot_beta_dist()
```

```python
a = 1
b = 1

N = 0
z = 0

plot_beta_dist(z + a, N - z + b, label='prior')

N += 1
z += 0
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z))

N += 5
z += 1
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z))

N += 10
z += 3
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z))


N += 100
z += 23
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z))

N += 1000
z += 235
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z), lw=5)

plt.legend();
```

```python
plot_beta_dist(z + a, N - z + b, label='N={}, z={}'.format(N, z))
plt.axvline(.24, ls='--', color='k')
plt.ylim(0, 50)
plt.xlim(.1)
```
