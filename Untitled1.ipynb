```python
df = pd.read_csv('censor.dat', delim_whitespace=True)
```

```python
plt.scatter(df.Teff, df.logN_Be);
```

```python
X = df.Teff
Y = df.logN_Be
```

```python
linear_model = pm.Model()

with linear_model:
    alpha = pm.Normal('Intercept', mu=0, sd=10) 
    beta = pm.Normal('Slope', mu=0, sd=10)
    sigma = pm.HalfNormal('noise_sd', sd=10)
    
    mean = alpha + beta*X 
    
    Y_obs = pm.Normal('Y_obs', mu=mean, sd=sigma, observed=Y)
```

```python
pm.find_MAP(model=linear_model)
```

```python
with linear_model:
    step = pm.Metropolis()
    trace = pm.sample(1000, step=step)
    
```

```python
pm.traceplot(trace);
```
