```python
import numpy as np
```

The `random` module in `numpy` conveniently has a Normal random number generator.  The `loc=` and `scale=` keyword arguments default to a mean of `0` and standard deviation of `1`, so we just need to specify that we want 1000 draws.

```python
draws = np.random.normal(size=1000)
```

Now lets make a histogram using `matplotlib`.

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(draws, density=True);
plt.xlabel('x')
plt.ylabel('$f(x|\mu=0, \sigma=1)$');
```

Now lets define a function for the probability density function (PDF) and plot it.

```python
xs = np.linspace(-5, 5, 100)
```

```python
def normal_pdf(x, mean=0, std=1):
    return 1/np.sqrt(2*np.pi*std**2) * np.exp(-(x - mean)**2/(2*std**2))
```

```python
normal_pdf(0)
```

```python
plt.hist(draws, density=True)
plt.plot(xs, normal_pdf(xs));
```

We could have also taken advantage of a predefined function in `scipy`.

```python
from scipy import stats
```

```python
a = stats.distributions.norm.pdf(xs)
```

```python
plt.plot(xs, a)
```
